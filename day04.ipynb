```python
#right angle triangle pattern using stars
n = int(input("enter the number:-"))
i = 1
while(i<=n):
    j = 1
    while(j<=i):
        print("*",end='')
        j=j+1
    print()
    i = i+1
        
```

```python
n = int(input("enter the number:-"))
i = 1
while(i<=n):
    k=(n-1)-i
    while(k>0):
        print(" ",end="")
        k=k-1
    j=i+1
    while(j>0):
       print("*",end='')
       j=j-1
    i=i+1
    print()
```
